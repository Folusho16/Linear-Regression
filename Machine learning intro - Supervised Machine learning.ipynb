```python
# we will explore the supervised machine learning first
# supervised machine learning is used to make predictions on an existing dataset
# supervised machine learning can predict numerical/continous values or probability of an event occuring


# we also have 2 types of supervised machine leaning
#1. Regression  - predicts continous/numerical values e.g age, salary, price, temperature
#2. Classification - predicts the probability of something happening e.g whether a customer will buy a product or not

import sklearn
```

```python
import pandas as pd

data = pd.read_csv("Salary_dataset.csv")
```

```python
data
```

```python
del data["Unnamed: 0"]
```

```python
data

#independent variable/column is YearsExperience
#dependent variable/column is Salary

# a simple supervised ml is when there is 1 independent and 1 dependent variable/column
```

```python
# plot the independent against the dependent column

import matplotlib.pyplot as plt

plt.scatter(data["YearsExperience"], data["Salary"])
plt.xlabel("Years of experience")
plt.ylabel("Salary")
```

```python
# store the independent and dependent column in a variable name
# x represents independent
# y represents dependent

x = data[["YearsExperience"]]  #x always have double bracket
y = data["Salary"]
```

# split x and y

```python
# we split x and y to get a better accuracy of the prediction

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)

#population and sampling

#train_test_split is the function that splits x and y
```

```python
len(x_test)
```

```python
#call the algorithm

from sklearn.linear_model import LinearRegression
```

# fitting the model

```python
model = LinearRegression().fit(x_train, y_train)

#train the model - machine is learning how to make prediction on its own
```

# making predictions

```python
model.predict(x_test).round()
```

# what will be the salary of an employee with 10 years experience
